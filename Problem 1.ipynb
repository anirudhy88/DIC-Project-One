```r
library(ggmap)
```

Used the above library to plot the map of usa based on the location of users
The below code just takes the latitude, longitude and administrative_area_level columns of the data frame
in the form of numeric values The administrative_area_level has been added as "map_state" into the existing data frame

```r
for(i in 1:nrow(dataFrameUserLoc)) {
    location<-geocode(as.character(dataFrameUserLoc$userLocation[i]), output = "more")
    if(!grepl(location[1],'NA')&"administrative_area_level_1" %in% names(location)&"lon" %in% names(location)&"lat" %in%  names(location)) {
        dataFrameUserLoc$lon[i] = as.numeric(location['lon'][1])
        dataFrameUserLoc$lat[i] = as.numeric(location['lat'][1])
        dataFrameUserLoc$map_state[i] = as.character(location['administrative_area_level_1'][[1]])
    }
    else {
        dataFrameUserLoc$lon[i] = 'NA'
        dataFrameUserLoc$lat[i] = 'NA'
        dataFrameUserLoc$map_state[i] = 'NA'
    }
}
```

```r
dataFrameUserLoc
```

```r
loca<-subset(dataFrameUserLoc,lon != 'NA' & lat != 'NA' & map_state != 'NA',)
```

Removing the rows from the data fram with 'NA' values

```r
loca
```

```r
install.packages("sqldf", repos="http://cran.us.r-project.org/")
```

```r
library(sqldf)
```

Now we group by the map_state column.

```r
final <- sqldf('select map_state, count(*) as frequency from loca group by map_state')
```

```r
final
```

Now, inorder to properly plot the locations, we need to get latitude and longitude of the states as below

```r
for(i in 1:nrow(final)) {
    finalCoord = geocode(final$map_state[i])
    final$lat[i] <- as.numeric(finalCoord[2])
    final$lon[i] <- as.numeric(finalCoord[1])
}
final
```

The below code is the actual implementation of plotting the final data onto the usa map.

```r
myMap <- get_map(location = 'United States', zoom = 4)
pointsPlot <- ggmap(myMap)
pointsPlot <- pointsPlot + geom_point(data = final, aes(size = frequency), color = 'red', alpha = 0.6)
pointsPlot <- pointsPlot + scale_size_area()
pointsPlot
```
